# 0-2. 크롤링 구글 드라이버

```python
# 크롬 창 열기
service = Service('chromedriver-win64/chromedriver-win64/chromedriver.exe')
driver = webdriver.Chrome(service=service)
driver.get('https://www.lookpin.co.kr/coordi')

# 페이지 소스 가져오기
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
img_tag = soup.find('img')
# 스크롤을 내리기 위한 액션 설정
actions = ActionChains(driver)

# 페이지를 여러 번 스크롤하여 추가 이미지를 로드
for i in range(120):  # 120번 스크롤 => 아마 천 장쯤?
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(0.1)  # 스크롤 후 페이지 로드 시간 대기

# 페이지 소스 가져오기
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

# 'sc-fPXMVe fCvpaB sc-cQCQeq jCtqUh' 클래스를 가진 div 태그 찾기
div_elements = driver.find_elements(By.CLASS_NAME, 'sc-fPXMVe.fCvpaB.sc-cQCQeq.jCtqUh')

# 이미지 저장 경로 설정
save_dir = r"C:/Users/wjdrl/OneDrive - dgu.ac.kr/데이터 분석/비어플/프로젝트 세션/2024_2학기/룩핀사진"

# img 태그의 src 속성을 추출하고 이미지 다운로드
for idx, div_element in enumerate(div_elements):
    try:
        # 해당 div 내부의 img 태그 찾기
        img_element = div_element.find_element(By.TAG_NAME, 'img')
        
        # img 태그에서 src 속성 가져오기
        img_url = img_element.get_attribute('src')
        
        # 이미지 URL 확인 및 다운로드
        if img_url:
            img_data = requests.get(img_url).content
            # 이미지 파일을 지정된 폴더에 저장
            img_path = os.path.join(save_dir, f'image_{idx}.jpg')
            with open(img_path, 'wb') as handler:
                handler.write(img_data)
            print(f"{img_path} 다운로드 완료")
        else:
            print(f"이미지 {idx}에서 src를 찾을 수 없습니다.")
    except Exception as e:
        print(f"이미지 {idx}에서 오류 발생: {e}")

# 웹 드라이버 종료
driver.quit()
```
